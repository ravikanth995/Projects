```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Exploratory Data Analytics**

```python
df.columns
```

```python
plt.figure(figsize = (20,5))

df1 = df.drop(35)

group_data = df1.groupby('States')['Population'].sum()
ax = group_data.plot(kind = 'bar')
totals  = []
for container in ax.containers:
  for bar in container:
    height = bar.get_height()
    totals.append(height)

total_sum = sum(totals)
for container in ax.containers:
  for bar in container:
    height = bar.get_height()
    per = height/total_sum*100
    ax.annotate(f'{per:.1f}%', xy = (bar.get_x()+bar.get_width()/2, height),
                xytext = (0,3), textcoords = 'offset points', ha = 'center',
                va = 'bottom', color = 'black', fontsize = 15)


plt.title("Indian Population 2011")
plt.ylabel("Population in Crores")
plt.grid(True)
plt.show()
```

# **Which State has largest in area ?**

```python
plt.figure(figsize = (20,6))

group_data = df1.groupby('States')['Area (km2)'].sum()

sorted_data = group_data.sort_values(ascending = True)

ax = sorted_data.plot(kind = 'bar', rot = 90)

totals  = []
for container in ax.containers:
  for bar in container:
    height = bar.get_height()
    totals.append(height)

total_sum = sum(totals)
for container in ax.containers:
  for bar in container:
    height = bar.get_height()
    per = height/total_sum*100
    ax.annotate(f'{per:.1f}%', xy = (bar.get_x()+bar.get_width()/2, height),
                xytext = (0,3), textcoords = 'offset points', ha = 'center',
                va = 'bottom', color = 'black', fontsize = 15)
plt.grid(True)
```

# **Conclusion**  

Population Distribution: Uttar Pradesh is the most populous state in India, followed by Maharashtra and Bihar. These three states have a significant share of the country's total population.

Regional Classification: Southern India, consisting of states like Karnataka, Andhra Pradesh, Tamil Nadu, Kerala, Telangana, and Puducherry, has a relatively lower population compared to states in Northern India, such as Uttar Pradesh and Bihar. This suggests a variation in population density across different regions.

Literacy Rates: Kerala has the highest literacy rate among the states, followed by Delhi and Uttarakhand. These states have made significant progress in education and literacy initiatives.

Gender Ratios: The states of Meghalaya, Kerala, and Tamil Nadu have relatively higher gender ratios, indicating a more balanced distribution between males and females. In contrast, states like Haryana and Punjab have lower gender ratios, suggesting a skewed gender distribution.

Population Growth: The states of Nagaland and Mizoram have experienced a negative population growth rate, indicating a decline in population over the given time period. On the other hand, states like Bihar, Uttar Pradesh, and Madhya Pradesh have shown relatively higher population growth rates.

Urbanization: States like Maharashtra, Tamil Nadu, and Gujarat have a higher concentration of urban population, as indicated by their significant urban population figures. This highlights the level of urban development and migration trends in these states.
