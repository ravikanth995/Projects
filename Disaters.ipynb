```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Natural_Disasters_in_India .csv')
df
```

```python
df.drop('Unnamed: 0', axis= 1, inplace = True)
df.head()
```

```python
df.isna().sum()
```

```python
df['Year'].value_counts()
```

```python
df.info()
```

```python
df1 = df.groupby('Year').count()
df1
```

```python
year = df1.index
year
```

```python
disaster_count = df1['Disaster_Info']
disaster_count
```

```python
plt.figure(figsize = (20,6))
sns.set_theme()
ax = sns.barplot(x = year, y = disaster_count, palette = 'bright')


plt.bar_label(ax.containers[0])
plt.xticks(rotation = 90)
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'pink',
                      height =2000,
                      width = 2000
                     ).generate(str(df["Title"]))
plt.rcParams['figure.figsize'] = (12,12)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Titles for Disaster 1990-2001")
plt.show()
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'pink',
                      height =2000,
                      width = 2000
                     ).generate(str(df["Disaster_Info"]))
plt.rcParams['figure.figsize'] = (12,12)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Reasons for Disaster in India 1990-2021")
plt.show()
```
