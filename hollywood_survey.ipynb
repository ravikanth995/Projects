# **Bar Plot **

**Is there any Flactuation in the Movie Growth over the Years ?**

```python
fig, ax = plt.subplots(figsize = (12,7))
df.groupby('Year')['Growth'].sum().plot(kind = 'bar', width = 0.9)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.grid(True)
```

Yes. There is Flactuation in the Growth of the Movie over the years. 
in the 1981 and 2020, due to COVID Pandemic there was huge decline in the Growth of the Movies of Hollywood Movies about -81% decline. But After, pandemic overed, we huge growth of the Movies in 2021 about 112% due to Opening of Market and Cinema theatre and Multiplexes for common man.

**Which Movie had the highest Growth ?**

```python
fig, ax = plt.subplots(figsize = (12,7))
df.groupby('#1 Release')['Growth'].sum().plot(kind = 'bar', width = 0.9)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.grid(True)
```

There is huge percentage change(GROWTH) in the of the 2020 released movie 'Bad Boys for Life' due to its reason that it was released during COVID Pandemic Lockdown world-wide and Prior to Lockdown of Pandemic 'It was Avendgers : EndGame' movie but due lockdown world wide, its collection was influenced.
StarWars had the highest Growth in the world.

**Which Movie Had The Highest Collection ?**

```python
fig, ax = plt.subplots(figsize = (12,7))
df.groupby('#1 Release')['Average'].sum().plot(kind = 'bar', width = 0.9)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Movies with Average Collections")
plt.grid(True)
```

Grease(1978) Average Collection is highest among the other movies followed by Star Wars Episode.

**How Many Movies Were Released Over the Years**

```python
plt.figure(figsize=(20, 8))
ax = sns.barplot(x='Year', y='Releases', data=df)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.xticks(rotation=90)
plt.xlabel('Years')
plt.ylabel('Releases')
plt.title('Stack Bar : Movies Released Over The Years')
plt.grid(True)
plt.show()
```

From the above observation, we observe that the highest number of movies released in 2018 declined by 83 out of about 910. Since the pandemic affected the global Lockdown, movies released were just 456 in 2020 and further decreased to 440 in 2021 but in 2022, there was jump by 56 releases of movies in 2022, after markets were  opened after Pandemic Lockdown.

**Pair Plot :** Relations among Year, Gross, Movie Releases and its Averages

```python
numerical_columns = ['Year', 'Total Gross', 'Growth', 'Average']
sns.pairplot(df, vars=numerical_columns, hue = 'Growth')
plt.title('Pairwise Relationships between Year, Total Gross, Growth and Average')
plt.show()
```

**Conclusion** Hollywood Movies had tremendous Gross Collection with respect to the Movie released over the years but was affected by the COVID19 Pandemic in 2020 but after Pandemic Lockdown was opened, from then it is thriving in the Movie releases and Gross Collection over the World due to its Content and Sci-fi Movies and Visual Effects.
