```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_html('https://wcdhry.gov.in/dashboard/beti-bachao-beti-padhao-bbbp/')
df[0]
```

```python
df = pd.DataFrame(data = df[0])
df
```

```python
df.drop('Sr. No', axis =1, inplace = True)
```

```python
df.rename(columns = {'Year Nov,2020':'Year 2020'}, inplace  = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.boxplot(x = 'Year 2016', data = df)
sns.boxplot(x = 'Year 2017', data = df)
sns.boxplot(x = 'Year 2018', data = df)
sns.boxplot(x = 'Year 2019', data = df)
sns.boxplot(x = 'Year 2020', data = df)

plt.grid(True)

plt.title("Boxplot : To determine the Outliers")
```

The Outliers can be observed in the above graph. 
There are some outliers in the box graph which represents some,
data collection errors, measurement errors, sampling issues, or genuine extreme values.

Outliers can result from various sources, including data entry mistakes, faulty equipment, rare events, or unique circumstances.



Exploratory Data Anaysis

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize = (20,5))
ax = sns.barplot(x = 'District', y = 'Year 2015', data = df, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Haryana State Data for 2015 of Girls Birth per 1000 Male Birth")
plt.xticks(rotation = 15)
```

In 2015, the number of girl birth per 1000 Male birth was the lowest in India.
This year, our honorable Prime Minister Narendra Modi Ji launched the Beti Bachao and Bati Padhao Scheme to balance the Gender Imbalance in the State with the State Government.

Was There Any Difference in the year 2016 after the Scheme was launched ?

```python
df['Diff'] = df['Year 2016'] - df['Year 2015']
plt.figure(figsize = (20,6))
ax = sns.barplot(x = 'District', y = 'Diff', data = df, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Girls  Birth Change in 2016 of Haryana State")
plt.xticks(rotation = 90)
```

The majority of the districts saw the positive impact of the Scheme.

```python
df['Diff1'] = df['Year 2017'] - df['Year 2015']
plt.figure(figsize = (20,6))
ax = sns.barplot(x = 'District', y  = 'Diff1', data = df, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.xticks(rotation = 90)
plt.title("Difference to the year 2015 in 2017 after the Scheme was launched")
plt.grid(True)
```

There was a great impact of Beti Bachao, Beti Padhao scheme in 2016. Yamunanagar District saw the Highest jump inn the Girls Birth with 75 Increase followed by the Rewari, Sonepet, Jhajjar.

```python
df['Diff2'] = df['Year 2018'] - df['Year 2015']

plt.figure(figsize = (20,6))

ax = sns.barplot(x = 'District', y = 'Diff2', data = df, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Difference in 2018 after the 2015 of Girls Birth")

plt.xticks(rotation = 90)
```

Jhajjar District saw the highest growth in the Girl's Birth with 123 and Narnaul with 100 because this was the place where Scheme was launched 

```python
df['Diff3'] = df['Year 2019'] - df['Year 2015']

plt.figure(figsize = (20,7))

ax = sns.barplot(x = 'District', y = 'Diff3', data = df, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Difference in the Girls Birth in the 2019")

plt.xticks(rotation = 90)
```

Narnaul, Rewari, jind and Ambala Districts were performed very well with 96,92,86 and 86.

```python
df['Diff4'] = df['Year 2020'] - df['Year 2015']

plt.figure(figsize = (20,7))

ax = sns.barplot(x = 'District', y = 'Diff4', data = df, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Difference in the Girls Birth in the 2020")

plt.xticks(rotation = 90)
```

Rewari District has performed consistently well in the Girl's Birth in 2020 but all districts have performed well in the girl's birth and there was no negative birth rate in 2020 as we had seen earlier
