```python
import tkinter as tk
from tkinter import *
from tkinter import filedialog
from tkinter.ttk import Combobox
import pyttsx3
import PyPDF2
import os



root=Tk()
root.title("Text/PDF to Speech")
root.maxsize(900,300)
root.minsize(900,300)
#color of the window
root.config(bg="#305065")

def speaknow():
    speaker=pyttsx3.init()
    text=text_area.get(1.0,END)
    gender=gender_box.get()
    speed=speed_box.get()
    voices=speaker.getProperty("voices")

    def setvoice():
        if (gender=="Female"):
            speaker.setProperty("voice",voices[1].id)
            speaker.say(text)
            speaker.runAndWait()
        else:
            speaker.setProperty("voice",voices[0].id)
            speaker.say(text)
            speaker.runAndWait()
    if (text):
        if (speed=="Fast"):
            speaker.setProperty("rate",250)
            setvoice()
        elif (speed=="Normal"):
            speaker.setProperty("rate",150)
            setvoice()
        else:
            speaker.setProperty("rate",100)
            setvoice()

def download():
    speaker=pyttsx3.init()
    text=text_area.get(1.0,END)
    gender=gender_box.get()
    speed=speed_box.get()
    voices=speaker.getProperty("voices")

    def setvoice():
        if (gender=="Female"):
            speaker.setProperty("voice",voices[1].id)
            path=filedialog.askdirectory()
            os.chdir(path)
            speaker.save_to_file(text,"text.mp3")
            speaker.runAndWait()
        else:
            speaker.setProperty("voice",voices[0].id)
            path=filedialog.askdirectory()
            os.chdir(path)
            speaker.save_to_file(text,"text.pm3")
            speaker.runAndWait()
    if (text):
        if (speed=="Fast"):
            speaker.setProperty("rate",250)
            setvoice()
        elif (speed=="Normal"):
            speaker.setProperty("rate",150)
            setvoice()
        else:
            speaker.setProperty("rate",100)
            setvoice()

def Upload():
    speaker=pyttsx3.init()
    text=text_area.get(1.0,END)
    gender=gender_box.get()
    speed=speed_box.get()
    voices=speaker.getProperty("voices")
    file_path = filedialog.askopenfilename(filetypes=[("PDF Files", "*.pdf")])

    def setvoice():
        speaker=pyttsx3.init()
        with open(file_path,"rb"):
            reader=PyPDF2.PdfReader(file_path)
            numberPage=len(reader.pages)
            for i in range(numberPage):
               page=reader.pages[i]
               text=page.extract_text()
               
        if (gender=="Female"):
            speaker.setProperty("voice",voices[1].id)
            
            speaker.say(text)
            speaker.runAndWait()
        else:
            speaker.setProperty("voice",voices[0].id)
           
            speaker.say(text)
            speaker.runAndWait()
    if (text):
        if (speed=="Fast"):
            speaker.setProperty("rate",250)
            setvoice()
        elif (speed=="Normal"):
            speaker.setProperty("rate",140)
            setvoice()
        else:
            speaker.setProperty("rate",100)
            setvoice()

#Icon
root.iconbitmap("iconmic.ico")

#Top frame
top_frame=Frame(root,bg="white",width=900,height=100)
top_frame.place(x=0,y=0)


Label(top_frame,text="TEXT/PDF TO SPEECH/PDF Viewer",bg="white",fg="black").place(x=100,y=10)

#text area
text_area=Text(root,font="Robote 20",bg="white",relief=GROOVE,wrap=WORD)
text_area.place(x=0,y=101,width=500,height=140)

Label(root,text="Voice",font="arial 15 bold", bg="#305065",fg="white").place(x=520,y=100)
Label(root,text="Speed",font="arial 15 bold", bg="#305065",fg="white").place(x=690,y=100)

#Gender Box
gender_box=Combobox(root,values=["Male","Female"],font="arial 14",state="r",width=7)
gender_box.place(x=520,y=130)
gender_box.set("Female")

#Speed box
speed_box=Combobox(root,values=["Fast","Normal","Slow"],font="arial 14",state="r",width=7)
speed_box.place(x=690,y=130)
speed_box.set("Normal")

#button
btn=Button(root,text="Speak",width=8,font="arial 14 bold",command=speaknow,bg="grey")
btn.place(x=520,y=170)

save=Button(root,text="Download",width=9,font="arial 14 bold",bg="#39c790",command=download)
save.place(x=690,y=170)
# 
browse=Button(root,text="Browse File",width=10,font="arial 14 bold",bg="#7F525D",command=Upload)
browse.place(x=520,y=220)

exitbtn=Button(root,text="Exit",bg="#FBB917",width=6,font="arial 14 bold",command=lambda:exit())
exitbtn.place(x=690,y=220)

root.mainloop()
```
