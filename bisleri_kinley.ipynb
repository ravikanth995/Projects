```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

<h1>In which Country Bisleri trends </h1>

```python
pytrend = TrendReq()

pytrend.build_payload(kw_list=['Bisleri'])
# Interest by Region
df = pytrend.interest_by_region()
df = df[df['Bisleri']>=1]

ax = df.reset_index().plot(x='geoName', y='Bisleri', figsize=(20, 5), kind ='bar')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')

plt.title("Bisleri Bottle Trend")
plt.show()
```

Bilseri trended in India, Pakistan, UAE, Italy, Bangladesh and Other Countries.

<h1>In which Country Kinley Trended ?</h1>

```python
pytrend = TrendReq()

pytrend.build_payload(kw_list=['Kinley'])
# Interest by Region
df = pytrend.interest_by_region()
df = df[df['Kinley']>=1]

ax = df.reset_index().plot(x='geoName', y='Kinley', figsize=(20, 5), kind ='bar')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')

plt.title("Kinley Bottle Trend")
plt.show()
```

Bhutan, Hungary, Slovakia and Bangladesh trended with Kinley.

```python
pytrends = TrendReq(hl='en-US', tz=360)

keywords = ['Bisleri']
pytrends.build_payload(keywords, cat=0, timeframe='today 5-y', geo='IN')

data = pytrends.interest_by_region(resolution='REGION', inc_low_vol=True, inc_geo_code=True)

# Print the data for all states
# print(data)

df1 = pd.DataFrame(data = data)
df1.shape
```

```python
pytrends = TrendReq(hl='en-US', tz=360)

keywords = ['Kinley']
pytrends.build_payload(keywords, cat=0, timeframe='today 5-y', geo='IN')

data = pytrends.interest_by_region(resolution='REGION', inc_low_vol=True, inc_geo_code=True)

# Print the data for all states
# print(data)

df2 = pd.DataFrame(data = data)
df.shape
```

```python
df = pd.merge(df1,df2)
df.style
```

```python

label1 = 'Bisleri'
label2 = 'Kinely'

color1 = 'magenta'
color2 = 'indigo'

ax = df.reset_index().plot(x='geoCode', y='Bisleri', figsize=(20, 7), kind='bar',label = label1, color = color1)
ax1 = df.reset_index().plot(x='geoCode', y='Kinley', kind='line', ax =ax,label = label2,color = color2)

label1 = 'Bisleri'
label2 = 'Kinely'
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
for i, row in df.iterrows():
    ax1.annotate(str(row['Kinley']), (i, row['Kinley']), xytext=(0, 10), textcoords='offset points', ha='center')


plt.xticks(rotation = 90)
plt.show()
```

```python
pytrend = TrendReq()

pytrend.build_payload(kw_list=['Bisleri'], timeframe='today 3-m')
df_bisleri_month = pytrend.interest_over_time()

pytrend.build_payload(kw_list=['Kinley'], timeframe='today 3-m')
df_kinley_month = pytrend.interest_over_time()

df_merged_month = pd.merge(df_bisleri_month, df_kinley_month, left_index=True, right_index=True)

ax_month = df_merged_month.plot(figsize=(20, 5))

for i, row in df_merged_month.iterrows():
    bisleri_val = row['Bisleri']
    kinley_val = row['Kinley']
    plt.text(i, bisleri_val, str(bisleri_val), ha='center', va='bottom')
    plt.text(i, kinley_val, str(kinley_val), ha='center', va='bottom')


ax_month.set_title('Trends by Month')
ax_month.set_xlabel("Month (in date)")
plt.show()
```
