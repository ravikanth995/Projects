# **Exploratory Data Analysis**

```python
df.info()
```

```python
plt.figure(figsize = (20,5))
group_data = df.groupby('Zone')['Order Amount'].mean()
plt.style.use('ggplot')
colors = ['orange','green','red','blue']
ax = group_data.plot(kind = 'bar', color = colors)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Average Amount Spend in Each Zone")
plt.xlabel("Zones")
plt.ylabel("Average Amount")
plt.grid(True)
plt.show()
```

In the Zone C, The Average Amount is more than followed by the Group D, Group A and B

```python
plt.figure(figsize = (20,5))
group_data = df.groupby('Zone')['Order Amount'].sum()
plt.style.use('ggplot')
colors = ['orange','green','red','blue']
ax = group_data.plot(kind = 'bar', color = colors)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Total of Amount Spend in Each Zone")
plt.xlabel("Zones")
plt.ylabel("Total Amount")
plt.grid(True)
plt.show()
```

Total Amount Spend in Zone is Highest in D, followed by Zone B, Zone A and Least Total Sum is Spend in C.

```python
df.info()
```

```python
total_cases_by_state = df.groupby('Quantity of Items')['Order Amount'].sum().reset_index()
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Quantity of Items', y='Order Amount', data=total_cases_by_state)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.xticks(rotation=90)
plt.xlabel('Quantity of Items')
plt.ylabel('Order Amount')
plt.title('Quantity of Items and Amount (Aggregated)')
plt.grid(True)
plt.show()
```

Total Amount spend on 5 Quantity is superseded other Items.

```python
df.columns
```

```python
total_sum = group_data.sum()
percentage_data = group_data / total_sum * 100

ax = percentage_data.plot(kind='bar', color=colors)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', label_type='edge')
plt.title("Average Amount in % paid v/s Payement Modes")
plt.xlabel("Payment Modes")
plt.ylabel("Amount in %")
plt.xticks(rotation = 20)
plt.grid(True)    
plt.show()
```

People prefer to pay the amount via Debit card but there are some slight differences in Payment Modes.

```python
ax = sns.barplot(data=df, x="Order Amount", y="Payment Mode", hue="Zone", palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'center')
plt.bar_label(ax.containers[1], fmt = '%d', label_type = 'center')
plt.bar_label(ax.containers[2], fmt = '%d', label_type = 'center')
plt.bar_label(ax.containers[3], fmt = '%d', label_type = 'center')

```

Payment Via Debit Card is Highest in Zone C.
Payment Via COD is Highest in Zone D.
Payment Via Credit Card is Highest in Zone A.

```python
df['Customer Rating-Food'] = df['Customer Rating-Food'].astype(float)
```
