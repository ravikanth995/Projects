<h1>Which Gender Contributes to Revenue Collection ?</h1>

```python
data_group = df.groupby('Customer_Gender')['Revenue'].sum()
colors = ['magenta','indigo']

ax = data_group.plot(kind = 'bar', alpha = 0.8, color = colors)
plt.grid(True)

totals = []
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        totals.append(height)

total_sum = sum(totals)

for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        percentage = height / total_sum * 100
        ax.annotate(f'{percentage:.1f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points', ha='center', va='bottom', color='black', fontsize=16)

plt.ylabel("Revenue")
plt.show()
```

Male Contributes to more revenue Collection than Female about 50.8% and 49.2%.
