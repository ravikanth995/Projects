```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/2011_Census_of_India')
data[7]
```

```python
df = pd.DataFrame(data = data[7])
df.head()
```

```python
df.rename(columns = {"% of total popula- tion[40]":"% of total population", "Rural[42] Population":"Rural Population", "Urban[42] Population":"Urban Population","Area[43] (km2)":"Area (km2)"}, inplace  =True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('Type', inplace = True, axis =1)
df.head()
```

```python
df.rename(columns = {'Sex Ratio [41]':'Sex Ratio'}, inplace = True)
df.columns
```

```python
df['Density (1/km2)'].value_counts()
```

```python
df.isna().sum()
df.dropna()
```

```python
df.head()
df.rename(columns = {"State / Union Territory (UT)":"States"}, inplace  = True)
df
```

```python
df1 = df[df['States'] == 'India']
df1
```

<h1>Exploratory Data Analysis</h1>

```python
plt.figure(figsize = (20,7))

df1 = df.drop(35)

group_data = df1.groupby('States')['Population'].sum()
ax = group_data.plot(kind = 'bar')
totals  = []
for container in ax.containers:
  for bar in container:
    height = bar.get_height()
    totals.append(height)

total_sum = sum(totals)
for container in ax.containers:
  for bar in container:
    height = bar.get_height()
    per = height/total_sum*100
    ax.annotate(f'{per:.1f}%', xy = (bar.get_x()+bar.get_width()/2, height),
                xytext = (0,3), textcoords = 'offset points', ha = 'center', 
                va = 'bottom', color = 'black', fontsize = 15)


plt.title("Indian Population 2011")
plt.ylabel("Population in Crores")
plt.grid(True)
plt.show()
```
