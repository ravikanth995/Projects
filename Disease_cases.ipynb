<h1>Exploratory Data Analysis of the World</h1>

#<h1>Malarial cases by Countries<h1>

```python
sum_val = df.groupby('Country')['Malaria'].sum()
data_group = sum_val[sum_val >= 10000]
plt.figure(figsize = (20,6))

ax = data_group.plot(kind = 'bar')
total_sum = data_group.sum()
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        percentage = height / total_sum * 100
        ax.annotate(f'{percentage:.2f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')
plt.title("Countries by Malarial cases from 1990-2020", fontsize = 20)
plt.grid(True)  
```

Nigeria, the Democratic Republic of Congo, India, and Uganda had the most Malarial Cases in the World
It might be due to the Geographical Location of the Countries which is comfortable for the mosquitos' growth.
