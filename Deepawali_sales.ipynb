```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame( data = df)
df
```

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.info()
```

# **Exploratory Data** **Analysis**

**Gender** **:**  plotting a bar chart for Gender and it's count


```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df, width = 0.8)
for bars in ax.containers:
  ax.bar_label(bars)
  plt.grid(True)
```

**Bar Chart** : Gender v/s Amount

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

**Summary :**
Female has Heighest Expenditure for Diwali Festival Than Men

# **Age**

```python

ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

**Summary** : From the Above Bar graph, it is evident that Female's between the 26-35 Age-Group has the highest Purchase Power than Men of the same age group and also other Age Groups too.

# **Total Amount vs Age Group**

What Age Group Buys More ??

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

**Summary :** From the above Bar graph, we can observe that Age Group between 26-35 wether Female or Male, has the highest Purchase power followed by 36-45 Age group. It diminishes after 46 Age.
the Teen age to Male Adult between 18-25 locates at 3rd After 36-25 might be due they are Students or Seeking Jobs.

# **State-Wise Purchase**

Which State had the Highest Orders ??

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders', alpha = 0.9)
plt.grid(True)
plt.xticks(rotation = 15)
```

From the above observation, we bring some summary: Uttara Pradesh State has the Highest Purchase during the Deepawali festival followed by the Maharashtra, Karnataka and Delhi.
Population and Income Per Capita also contributes to the Products Sale during Festival seasons.
So, sellers must focus on the Population and IPC(Income Per Capita) of the States for their Products Sell.

**Hypotheis :** Does Marital Status Also Contribute the Products sells During Festival ??

```python

ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

From the Above Graph Obersvation : We can really observe that, Unmarried people tend to spend to more income on Festival Purchase than Married people. 
**Reason** : Married Couples tend to save thier income than Unmarried people. 

Hypothesis : Does Occupation or Vocation Contributes to Purchase during the Festival Seasons ??

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

The Graph explains the Which sector Occupants spend more : The IT Sector Spends more on Festival Season, followed by the Healthcare sector, Aviation and Banking Sector.
**Reason** : IT sector offers high income than other sectors.

Which Sector has the highest Buyer ??

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

IT Sector has the highest Buyer followed by Healthcare, Aviation and Banking.

Which Products had the Highet Orders ??

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
    plt.xticks(rotation = 15)
```

Clothing & Apparel had the Highest Order from Customer followed by the Food and Electronic & Gadgets.

Which Product had the highest Revenue ?

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

Products Based on Foods had the Highest Revenue followed by the Clothing & Apparel and Electronic & Gadgets.
