# **Exploratory Data Analysis**

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize = (10,3))
sns.countplot(x = 'Label', data = df)
```

Moisturizer is most in number followed by the Cleanser, Treatment, Facemask, Eye Cream and Sun Protect.

```python

plt.figure(figsize=(25, 8))
data_group = df.groupby('Brand')['Price'].max()

data_group = data_group[data_group >= 100]

ax = data_group.plot(kind='bar')
plt.bar_label(ax.containers[0], fmt='%s', label_type='edge')
plt.show()
```

These are the Cosmetic Products who are above 100 Rupees in INR. La MER stands 1st spot on the Price of 370 followed by SHISEIDO.

```python
df.columns
```

```python
group_da = df.groupby('Label')['Price'].sum()
colors = ['indigo','magenta','orange','blue','Teal','Coral']
ax = group_da.plot(kind = 'bar', color = colors)
plt.bar_label(ax.containers[0], fmt='%s', label_type='edge')
plt.title("Cosmetics Types v/s Prices")
plt.ylabel("Price")
plt.xlabel("Types of Cosmetics")
plt.show()
```

Moisturizer is more expenstive than Treatement, Eye Cream, Cleanser and Sun Protect and Face Mask.

```python
df.columns
```

```python
group_da = df.groupby('Label')['Dry'].sum()
colors = ['indigo','magenta','orange','blue','Teal','Coral']
ax = group_da.plot(kind = 'bar', color = colors)
plt.bar_label(ax.containers[0], fmt='%s', label_type='edge')
plt.title("Cosmetics Types v/s Prices")
plt.ylabel("Price")
plt.xlabel("Types of Cosmetics")
plt.show()
```

Moisturizer is the most expenstive followed by the Face mask, Treatment, Eye Cream and Cleanser.

```python
data_group  = df.groupby('Label')['Combination'].sum()
explode = [0.1]*6
data_group.plot(kind = 'pie', autopct = '%.2f%%', explode = explode, shadow = True)
plt.show()
```

Face Mask had the highest combination of Dry and Oily Skin Effects with 20.91%, Moisturizer had with 20.60% and followed by the Treatment with 18.32%, Eye Cream had 13.56% and Least Combination of Oily and Dry Effects on Skin were Sun Protect.
