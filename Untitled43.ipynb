<h1>Exploratory Data Analysis</h1>

```python
df1  = df[df['Department'] == 'Driving and Transport Business']
df1.style
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

max_row = df.loc[df['Amount(INR)'].idxmax()]

max_date = max_row['Date']

print('Maximum Amount(INR) occurred on:', max_date)
```

```python
max_row = df.loc[df['Amount(INR)'].idxmin()]

max_date = max_row['Date']

print('Minimum Amount(INR) occurred on:', max_date)
```

```python
filtered_df = df[df['Location'] == 'Bangalore,Karnataka']


if filtered_df.empty:
    print("No rows match the filter condition.")
else:
    filtered_data = filtered_df[filtered_df['Amount(INR)'] > 100000]
    ax = filtered_data.plot(x='Date', y='Amount(INR)', kind='bar')
    plt.bar_label(ax.containers[0], fmt = '%d', label_type= 'edge')

    plt.xlabel('Date')
    plt.ylabel('Amount in Indian Rupees')
    plt.title("Bangaluru")
```

```python
plt.figure(figsize = (20,5))
filtered_df = df[df['Location'] == 'Bangalore,Karnataka']
if filtered_df.empty:
    print("No rows match the filter condition.")
else:
    filtered_data = filtered_df[filtered_df['Amount(INR)'] >10000]
    group = filtered_data.groupby('Department')['Amount(INR)'].sum()
    group.plot(kind = 'bar')

  

    plt.xlabel('Date')
    plt.ylabel('Amount in Indian Rupees')
    plt.title("Bangaluru")

```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'pink',
                      height =2000,
                      width = 2000
                     ).generate(str(df["Title"]))
plt.rcParams['figure.figsize'] = (12,12)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Motives for Corruptions ")
plt.show()
```

```python
stopwords = set(STOPWORDS)
filtered_df = df[df['Location'] == 'Bangalore,Karnataka']  # Filter the DataFrame for 'Bangalore, Karnataka'

wordcloud = WordCloud(background_color='pink',
                      height=2000,
                      width=2000,
                      stopwords=stopwords  # Use stopwords to remove common words
                     ).generate(str(filtered_df["Title"]))  # Generate the Word Cloud for the filtered data

plt.rcParams['figure.figsize'] = (12, 12)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Motives for Corruptions in Bangalore, Karnataka")  # Add the location to the title
plt.show()
```
