ON AN AVERAGE, HOW MUCH AN OCCUPATION FETCHES INCOME.

```python
mean_income_by_occupation = df.groupby('Occupation')['Monthly Income'].mean()

# Plotting a bar chart of the mean income by occupation
plt.style.use('ggplot')
color = ['red', 'blue', 'orange', 'green']
ax = mean_income_by_occupation.plot(kind='bar', color = color)
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title('Occupation v/s Income')
plt.xlabel('Occupation')
plt.ylabel('Mean Monthly Income')
plt.grid(True)
plt.show()
```

On an Average, Self-employed earns more income than Employed followed by Students.

Analysing Outliers in the Data

```python
# Box plot to visualize outliers in a numerical variable
import numpy as np
sns.boxplot(x='Age', data=df)
q = df['Age'].quantile(0.99)
df['Age'] = np.where(df['Age'] > q, q, df['Age'])
plt.show()

```

Outliers are on the edges of 18 and 30-32.
Outliers : In this context of dataset, "outliers" refer to data points that deviate significantly from the overall pattern or distribution of this dataset.

Outliers can have a significant impact on data analysis and statistical modeling. They can distort statistical measures such as the mean and standard deviation, leading to misleading results. Therefore, it is important to identify and handle outliers appropriately in order to obtain accurate and reliable insights from this dataset.
