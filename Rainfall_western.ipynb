# ***Exploratory Data Analysis***

```python
df['SUBDIVISION'] = df['SUBDIVISION'].str.replace('MATATHWADA','MARATHWADA')
```

```python
row = df.loc[df['SUBDIVISION'] == 'MARATHWADA']
rows = row.index
rows
```

```python
column_to_plot = df.loc[2737:2851,'Jun-Sep']
column_to_plot
```

```python
columns = df.loc[(df['YEAR'].between(2000, 2015)) & (df.index.isin(rows))]
```

```python
df.columns
```

```python

plt.figure(figsize = (25,8))
plt.style.use('fivethirtyeight')
ax = sns.barplot(data=columns, x='YEAR', y=column_to_plot, palette = 'bright')
plt.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge')
plt.title("Rainfall in June-September in MARATHWADA REGION OF MAHARASHTRA 2000-2015 in mm")
plt.grid(True)
```

```python
sorted_data = columns.sort_values('YEAR')

plt.figure(figsize = (25,8))
plt.style.use('fivethirtyeight')
sns.lineplot(data=columns, x='YEAR', y=column_to_plot)

for x, y in zip(sorted_data['YEAR'], column_to_plot):
    plt.annotate(text=str(y), xy=(x, y), xytext=(0, -20), textcoords='offset points', ha='center', fontsize=20)


plt.title("Rainfall in June-September in MARATHWADA REGION OF MAHARASHTRA 2000-2015 in mm")
plt.grid(True)
```

There is consistently drop in the rainfall over Marathwada region of Maharashtra due to,
1. Distant from the Sea.
2. No river or lake in such quantity that, it can feed people in the region.
3. Non-connectivity of water pipelines in the region from municipality.
4. Unscientific calculation or ill-scientific advices to the government about the water conservation planning.
