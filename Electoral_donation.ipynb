```python
group_data = df.groupby('Financial Year ')['Amount'].sum()

group_data
```

<h1>Maximum Donations to a Political Party from 2004-2015</h1>

```python
ax = group_data.plot(kind = 'bar')
total = group_data.sum()
for container in ax.containers:
    ax.bar_label(container, labels=[f'{x.get_height()/total*100:.2f}%' for x in container], label_type='edge')

plt.title("Amount Donation from 2004-2015", fontsize = 20)
plt.grid(True)
```

There is a jump of 2x in donation in amount to political party from 2013-2015.

```python
df.columns
```

```python
df1 = df[df['Financial Year '] == '2014-15']
df1[['Party','PAN Given','Amount']].value_counts().head(10)
```

<h1>Maximum Donations to Political Party in 2014-15</h1>

```python
ax = sns.barplot(x = 'Party', y = 'Amount', hue = 'Financial Year ',data = df1)
total = group_data.sum()
for container in ax.containers:
    ax.bar_label(container, labels=[f'{x.get_height()/total*100:.2f}%' for x in container], label_type='center')

plt.grid(True)
```

NCP Stands maximum donations in the Political Donations followed by the INC and BJP in 2014-15

<h1>Amount Increase in the Political Party in the Financial Years from 2005-15</h1>

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(20, 7))

ax = sns.lineplot(x='Financial Year ', y='Amount', hue='Party', data=df.sort_values('Financial Year ', ascending=True), palette = 'bright')

plt.grid(True)


plt.title("Amount Increase in Financial Year", fontsize = 20)
plt.grid(True)
```

Accodring to the Data Given, NCP has been the Political party to whom maximum donations has been given officially followed by the Congress and BJP.

```python
group_data = df1.groupby('Name')['Amount'].sum()

group_data
```

<h1>Which Person/Company Donated Maximum Amount to Which Political Party</h1>

```python
plt.style.use('default')
plt.figure(figsize=(23,6))
ax = sns.barplot(x='Name', y='Amount', data=df1[df1['Amount'] >= 16000000], hue = 'Party')
total = group_data.sum()
for container in ax.containers:
    ax.bar_label(container, labels=[f'{x.get_height()/total*100:.2f}%' for x in container], label_type='edge')

plt.xticks(rotation = 90)
```

General Electoral Trust, Satya Electoral Trust and Videocon Industries Ltd. most to the Political Party officially.

<h1>Donations via Which Mode ?</h1>

```python
plt.style.use('default')
plt.figure(figsize=(23,6))
ax = sns.barplot(x='Name', y='Amount', data=df1[df1['Amount'] >= 20000000])

total = group_data.sum()
for container in ax.containers:
    ax.bar_label(container, labels=[f'{x.get_height()/total*100:.2f}%' for x in container], label_type='edge')

plt.grid(True)
plt.xticks(rotation = 90)
```

Through the HDFC Bank donations are done mostly.
